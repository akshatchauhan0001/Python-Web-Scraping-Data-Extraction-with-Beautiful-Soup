<h1>Project 1 : Creating a ebay Price Tracker</h1>

```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.ebay.com/p/14056251738?iid=204743589468"
res=requests.get(url)
print(res)
```

```python
soup = BeautifulSoup (res.text, 'lxml')
unedited = soup. select (".x-bin-price__content .ux-textspans")
uneditedPrice = unedited[0]
uneditedPrice = uneditedPrice.text
print(uneditedPrice)
```

```python
formattedPrice = uneditedPrice[1:]
print(formattedPrice)
formattedPrice = formattedPrice.replace(",", "")
price= float(formattedPrice)
print(price)
```

```python
actual_price=float(formattedPrice)
myPrice = 800
if actual_price < myPrice:
  print("The price is below your budget, go grab them" )
else:
  print("Price of product is above your budget")
```

```python
import matplotlib.pyplot as plt
# Visualization
prices = [price, myPrice]
labels = ['Actual Price', 'Your Budget']
colors = ['red', 'green']

plt.bar(labels, prices, color=colors)
plt.ylabel('Price ($)')
plt.title('Price Comparison')
plt.show()
```
