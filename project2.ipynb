<h1>Project 2 : Obtaining the Top 20 Hits from BillBoard




```python
import requests
from bs4 import BeautifulSoup
url="https://www.billboard.com/charts/hot-100/"
res=requests.get(url)
print(res)
```

```python
soup = BeautifulSoup(res.text, 'lxml')
```

```python
songs= soup.select("li h3")
```

```python
top20 =[song.text[14:][0:-5] for song in songs[:20]]
print(top20)
```

```python
i = 0
for song in top20:
  i = i + 1
  print(f"{i}) {song}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
ranks = list(reversed(range(1, len(top20) + 1)))  # Convert the generator to a list
plt.barh(range(len(top20)), ranks, color='skyblue')
plt.yticks(range(len(top20)), reversed(top20), fontsize=10)
plt.xlabel('Ranking')
plt.title('Top 20 Billboard Hot 100 Songs')
plt.gca().invert_yaxis()  # To have the highest rank at the top
plt.show()
```
